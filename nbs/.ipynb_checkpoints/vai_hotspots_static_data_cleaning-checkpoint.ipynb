```python
pop_df = pd.read_excel("india_district-wise_census.xlsx")
```

```python
pop_df.columns
```

```python
population_dict = {}

for row in pop_df.itertuples():
    population_dict[getattr(row, "Name").strip().replace("Bangalore", "Bengaluru")] = getattr(row, "TOT_P")
```

```python
for dist, pop in population_dict.items():
    print(dist)
```

```python
df['District'].replace({
    'Kalaburagi': 'Gulbarga',
    'Bagalkote': 'Bagalkot',
    "Ballari": "Bellary",
    "S.P.S. Nellore": "Sri Potti Sriramulu Nellore",
    "Kanyakumari": "Kanniyakumari",
    'Kondagaon': 'Uttar Bastar Kanker',
    "Mahabubnagar": "Mahbubnagar",
    "Jagatsinghpur": "Jagatsinghapur",
    "Ahmednagar": "Ahmadnagar",
    "Jagtial": "Dharmapuri",
    "Buldhana": "Buldana",
    "Belagavi": "Belgaum",
    "Shivamogga": "Shimoga",
    "Chikkamagaluru": "Chikmagalur"
}, inplace=True)

districts_to_remove = ['Bengaluru Urban', 'Jangaon', 'Mancherial', 'Medchal Malkajgiri']

df = df[df.District.map(lambda x: x not in districts_to_remove)]
```
