```python
df = pd.read_csv(data + 'all_with_population.csv')
df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'Area (sq km)'], inplace=True)
df.dropna(subset = ["Population"], inplace=True)
df
```

```python
long = []
lat = []
for coords in df.geometry:    
    coord = coords.split(",")
    cmp = []
    for c in coord:
        temp = re.findall(r'\d+', c) 
        pt = np.array([temp[0]+'.'+temp[1], temp[2]+'.'+temp[3]], dtype=np.float64)
        cmp.append(pt)

    cmp = np.mean(np.array(cmp), axis=0)
    long.append(cmp[0])
    lat.append(cmp[1])
```

```python
df['Longitude'] = long
df['Latitude'] = lat
```

```python
df.Zone.replace(to_replace='Red', value=-1, inplace=True)
df.Zone.replace(to_replace='Green', value=1, inplace=True)
df.Zone.replace(to_replace='Orange', value=0, inplace=True)
df.head()
```

```python
df.to_csv("all_with_long_lat.csv")
```
